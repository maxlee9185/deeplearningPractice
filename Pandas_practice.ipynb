```python
import numpy as np

col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:',array1.shape)
#리스트를 이용해 DataFrame 생성
df_list1=pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n',df_list1)
#넘파이 ndrarray를 이용해 DataFrame 생성
df_array1=pd.DataFrame(array1,columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n',df_array1)
```

```python
col_name2=['이름','나이','고향']
list=[['이성현','27','성북'],['이진권','23','방배']]
array=np.array(list)
print('array shape: ',array.shape)
df_list=pd.DataFrame(list,columns=col_name2)
print()
print('2차원 리스트로 만든 DataFrame:\n',df_list)
df_array=pd.DataFrame(array,columns=col_name2)
print()
print('2차원 ndarray로 만든 DataFrame:\n',df_array)
```

```python
dict={'col1':[1,11],'col2':[2,22],'col3':[3,33]}
df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
```

```python
titanic_df['Family_No']=titanic_df['SibSp']+titanic_df['Parch']+1
titanic_df.head(3)
```

```python
titanic_df['Age_0']=0
titanic_df.head()
```

```python
titanic_df.drop('Age_0',axis=1)
```

```python
drop_result=titanic_df.drop(['Age_0','Family_No'],axis=1,inplace=True)
print('inplace=True로 변환후 drop',drop_result)
titanic_df.head()
```

```python
pd.set_option('display.width',1000)
pd.set_option('display.max_colwidth',15)
print('#### before axis 0 drop ####')
print(titanic_df.head(3))
titanic_df.drop([0,1,2],axis=0,inplace=True)
print('#### after axis 0 drop ####')
print(titanic_df.head(3))
```
